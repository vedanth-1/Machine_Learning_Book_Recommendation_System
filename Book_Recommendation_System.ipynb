```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
padhe_likhe_users = x[x].index
padhe_likhe_users
#padhe_likhe_users.shape
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
famous_books.shape
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt.columns
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = Book_data[Book_data['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        #item.extend(list(temp_df.drop_duplicates('User-ID')['User-ID'].values))
        
        data.append(item)
    
    return data
```

```python
from sklearn.metrics import average_precision_score
# Sample ground truth data (books that the user has interacted with)
ground_truth = ['1984', 'Brave New World', 'Animal Farm']
# Sample recommended books
recommended_books = ['Animal Farm', 'The Handmaid\'s Tale', '1984', 'Brave New World']
# Create binary relevance data (1 if recommended book is in ground truth, 0 otherwise)
y_true = [int(book in ground_truth) for book in recommended_books]
# Calculate the average precision
map_score = average_precision_score(y_true, range(1, len(y_true) + 1))
print(f"Mean Average Precision (MAP): {map_score:.2f}")
```

```python
from sklearn.metrics import recall_score, f1_score
# Sample ground truth data (books that the user has interacted with)
ground_truth = ['1984', 'Brave New World', 'Animal Farm']
# Sample recommended books
recommended_books = ['Animal Farm', 'The Handmaid\'s Tale', '1984', 'Brave New World']
# Create binary relevance data (1 if recommended book is in ground truth, 0 otherwise)
y_true = [int(book in ground_truth) for book in recommended_books]
# Calculate recall
recall = recall_score(y_true, [1] * len(y_true))  # Recall = TP / (TP + FN)
# Calculate F1 score
f1 = f1_score(y_true, [1] * len(y_true))  # F1 = 2 * (precision * recall) / (precision + recall)
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
# Initialize counts for each book
book_counts = {book_name: {'rated_high': 0, 'rated_low': 0, 'not_rated': 0} for book_name in recommended_books}

# Iterate through each user
for user_id in pt.columns:
    for book_name in recommended_books:
        # Check if the user has rated the book
        if pt.at[book_name, user_id] > 0:
            # Check the rating
            if pt.at[book_name, user_id] > 5:
                book_counts[book_name]['rated_high'] += 1
            else:
                book_counts[book_name]['rated_low'] += 1
        else:
            book_counts[book_name]['not_rated'] += 1

# Sort the books by the count of >5 ratings in descending order
sorted_books = sorted(book_counts.items(), key=lambda x: x[1]['rated_high'], reverse=True)

# Display the top 5 recommendations with counts
for i, (book_name, counts) in enumerate(sorted_books[:5]):
    print(f"Top Recommendation {i + 1}: {book_name}")
    print(f"Rated >5: {counts['rated_high']} users")
    print(f"Rated <5: {counts['rated_low']} users")
    print(f"Not rated: {counts['not_rated']} users")
    print()
```
